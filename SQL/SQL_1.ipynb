**Задание 4**

Сохраните таблицы в переменные

- Таблицу bookings в переменную df_bookings

- Таблицу tickets в переменную df_tickets

И так далее по всем таблицам

```python
df_bookings = pd.read_sql("SELECT * FROM bookings", con=conn)
df_bookings
```

```python
df_tickets = pd.read_sql("SELECT * FROM tickets", con=conn)
df_tickets
```

```python
df_ticket_flights = pd.read_sql("SELECT * FROM ticket_flights", con=conn)
df_ticket_flights
```

```python
df_flights = pd.read_sql("SELECT * FROM flights", con=conn)
df_flights
```

```python
df_airports = pd.read_sql("SELECT * FROM airports", con=conn)
df_airports
```

```python
df_aircrafts = pd.read_sql("SELECT * FROM aircrafts", con=conn)
df_aircrafts
```

```python
df_boarding_passes = pd.read_sql("SELECT * FROM boarding_passes", con=conn)
df_boarding_passes
```

```python
df_seats = pd.read_sql("SELECT * FROM seats", con=conn)
df_seats
```
